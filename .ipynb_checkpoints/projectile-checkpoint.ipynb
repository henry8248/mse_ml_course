### prepare dataset

```python
import pandas as pd
import numpy as np

g = 9.8
```

```python
def R(V_init, theta):
    return 2*(V_init**2)*np.cos(theta)*np.sin(theta)/g

def H(V_init, theta):
    return (V_init**2)*np.sin(theta)/(2*g)
```

```python
print(np.sin(30))
```

```python
random_V_init = np.random.uniform(0, 100, (1000,))
random_theta = np.random.uniform(0, np.pi/2, (1000,))
```

```python
print(random_V_init.shape)
print(random_theta.shape)
```

```python
R_dist = R(random_V_init, random_theta)
H_dist = H(random_V_init, random_theta)
```

```python
print(R_dist)
print("---------\n")
print(H_dist)
```

```python
## Combine the dataset as a whole picture
dataset = pd.DataFrame({"V":random_V_init, "theta":random_theta, "R":R_dist, "H":H_dist})
```

```python
dataset
```

```python
import tensorflow as tf
```

```python
print(tf.config.list_physical_devices("CPU"))
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X = dataset[['V', 'theta']]
Y = dataset[['R', 'H']]
```

```python
X, Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y , 
                                   random_state=104,  
                                   test_size=0.25,  
                                   shuffle=True) 
```

```python
import tensorflow as tf
```

### Build the model architecture

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(tf.keras.Input(shape=(2,)))
model.add(Dense(2, activation='relu'))
model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.Dense(2))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss="MAE",
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test, verbose = 1)
```
